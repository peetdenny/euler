```python
from matplotlib import pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import numpy as np
import math
from IPython.display import display, clear_output
```

```python
v = np.array([[5,5], [4,5], [-2,3]])
origin = [[0],[0]]
for theta in range(359,345,-1):  # degrees 1 -> 10
    figure = plt.quiver(*origin,v[:,0], v[:,1], color=['r','g','b','pink'], scale=21)
    plt.xlim(-10, 10)
    plt.ylim(-10, 10)
    plt.grid(b=True, which='major')
    display(figure)
    clear_output(wait = True)
    #m = np.array([[1,2],[3,1]])
    m = np.array([[math.cos(theta),math.sin(theta)],[- math.sin(theta),math.cos(theta)]])
    plt.pause(0.5)
    v = np.matmul(v,m)
```

```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from itertools import product, combinations

v = np.array([[0,0,5], [4,5,4], [-5,3,-2]])
fig = plt.figure()
ax = fig.gca(projection='3d')
ax.set_xlim3d(-100, 100)
ax.set_ylim3d(-100, 100)
ax.set_zlim3d(-100, 100)

ax.scatter(v[:,0], v[:,1], v[:,2], color="g", s=100)
display(ax)
clear_output(wait = True)
plt.pause(1)
fig = plt.figure()
ax = fig.gca(projection='3d')
ax.set_xlim3d(-100, 100)
ax.set_ylim3d(-100, 100)
ax.set_zlim3d(-100, 100)

m = np.array([[0,3,6],[1,4,7],[2,5,8]])
v = np.matmul(v,m)
ax.scatter(v[:,0], v[:,1], v[:,2], color="g", s=100)
display(ax)

```
